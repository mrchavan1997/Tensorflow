##Creating Tensors using Tensorflow and manipulating it.

```python
# Use of tf.constant()
```

```python
import tensorflow as tf
print(tf.__version__)
```

```python
scalar = tf.constant(7)
scalar
```

```python
scalar.ndim
```

```python
vector = tf.constant([3,5])
vector
```

```python
vector.ndim
```

```python
matrix = tf.constant([[3,4],[6,7]])
matrix
```

```python
matrix.ndim
```

```python
matrix_2 = tf.constant([[2,3],[4,6]], dtype = tf.float16) #float takes less space in RAM than int
matrix_2
```

```python
# Use of tf.Variable()
```

```python
changeable_tensor = tf.Variable([7,8])
unchangeable_tensor = tf.constant([9,8])
changeable_tensor, unchangeable_tensor
```

```python
changeable_tensor[0].assign(10)
```

```python
#we can't change the value of tf.constant

unchangeable_tensor[0].assign(34)
```

```python
# Random Tensors

randomT_1 = tf.random.Generator.from_seed(23)
randomT_1 = randomT_1.normal(shape = (3,3))

randomT_2 = tf.random.Generator.from_seed(23)
randomT_2 = randomT_2.normal(shape = (3,3))

randomT_1, randomT_2
```

```python
randomT_3 = tf.random.uniform(shape = (2,2),minval = 1, maxval = 10)
randomT_3
```

```python
# One hot encoding
```

```python
list = [0,1,2,3]

tf.one_hot(list, depth = 4)
```

```python
tf.one_hot(list, depth = 4, on_value = 'YES', off_value = 'NO')
```

```python
tf.one_hot(list, depth = 3)
```

```python
# Math operations
list  = [7,9,3,5]
list_tensor = tf.constant(list)
```

```python
tf.square(list_tensor)
```

```python
tf.sqrt(tf.cast(list_tensor, dtype = tf.float16))
```

```python
tf.math.log(tf.cast(list_tensor, dtype = tf.float16))
```

```python
tf.math.sigmoid(tf.cast(list_tensor, dtype = tf.float16))
```
