#In this Notebook we will create dummy tensors to solve the regression problem.

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.array([4,6,8,10,12,14,16])
y = np.array([2,4,6,8,10,12,14])
x,y
```

```python
x = tf.constant(tf.cast(x, dtype = tf.float16))
y = tf.constant(tf.cast(y, dtype = tf.float16))
x,y
```

```python
plt.scatter(x,y)
```

```python
y = x - 2 #We will train our Neural Network to figuree out this equation and predict the values.
```

```python
#Sequential model creation
```

```python
#Model Creation
model  = tf.keras.Sequential([
  tf.keras.layers.Dense(1)
])

#Model Compilation

model.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.SGD(),
    metrics = ['mae']
)

#fit the model

model.fit(x,y, epochs = 5)
```

```python
# So here we are using MAE as a loss function so we will train the model by changing the epochs values unless and until we get minimum MAE otherwise model will give wrong output
```

```python
model.predict([7])
```

```python
# Improving the Model

#Model Creation
model  = tf.keras.Sequential([
  tf.keras.layers.Dense(1)
])

#Model Compilation

model.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.SGD(),
    metrics = ['mae']
)

#fit the model

model.fit(x,y, epochs = 20)
```

```python
# here we can see the difference between the Loss of both models. Now lets try to predict the values.
```

```python
model.predict([18])
```

```python
model.predict([7])
```

### How to improve model
1. Try different epoch values
2. Add hidden layers
3. try different optimizers with different learning rate.

```python
model_1 = tf.keras.Sequential([
  tf.keras.layers.Dense(10, input_shape = [1]),
  tf.keras.layers.Dense(1)
])

model_1.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(), 
    metrics = ['mae']
)

model_1.summary()
```
